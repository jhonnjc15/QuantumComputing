```python
def oracle(input_reg,output_reg,qc):
    # Do not modify anything above this line
    
    qc.barrier(input_reg,output_reg)
    qc.cx(input_reg[2], output_reg[0])
    qc.cx(input_reg[1], output_reg[0])
    qc.barrier(input_reg,output_reg)

    return qc
input_reg = QuantumRegister(3)
output_reg = QuantumRegister(1)
creg = ClassicalRegister(1)
qc = QuantumCircuit(input_reg, output_reg, creg)
print(oracle(input_reg,output_reg,qc))
```

```python
#Comprobando el funcionamiento del oráculo
def test(input_state):
    input_reg = QuantumRegister(3)
    output_reg = QuantumRegister(1)
    creg = ClassicalRegister(1)
    qc = QuantumCircuit(input_reg, output_reg, creg)
    
    #initialization
    if input_state[0]=="1":
        qc.x(input_reg[2])
    if input_state[1]=="1":
        qc.x(input_reg[1])    
    if input_state[2]=="1":
        qc.x(input_reg[0])
    
    #call oracle
    oracle(input_reg,output_reg,qc)
    qc.measure(output_reg,creg)
    print(qc)
    job = execute(qc, Aer.get_backend('qasm_simulator'),shots=1)
    counts = job.result().get_counts()
    return counts

for s in ["001","000","111","110","011","010","100","101"]:
    print("Input:" ,s, "Output:", test(s), "\n\n")
```

```python
def oracle_part(qc,qr):
    mcx = XGate().control(3)
    
    #Los qubits del 0 al 4 son los nodos.
    #Los qubits del 5 al 10 nos permiten reconocer si la arista es correcta
    #(i.e. si se da entre nodos de distinto color).
    qc.cx(qr[0], qr[5])
    qc.cx(qr[3], qr[5])
    qc.barrier()
    qc.cx(qr[0], qr[6])
    qc.cx(qr[4], qr[6])
    qc.barrier()
    qc.cx(qr[1], qr[7])
    qc.cx(qr[3], qr[7])
    qc.barrier()
    qc.cx(qr[1], qr[8])
    qc.cx(qr[4], qr[8])
    qc.barrier()
    qc.cx(qr[2], qr[9])
    qc.cx(qr[3], qr[9])
    qc.barrier()
    qc.cx(qr[2], qr[10])
    qc.cx(qr[4], qr[10])
    qc.barrier()
    
    #Los qubits del 11 al 13 nos van a permitir contar la cantidad de aristas correctas
    #de tal forma que la suma será expresada en binario con cifras dadas por q_13, q_12 y q_11.
    qc.barrier()    
    qc.cx(qr[5],qr[11])
    qc.barrier()
    
    qc.ccx(qr[6], qr[11], qr[12])
    qc.cx(qr[6],qr[11])
    qc.barrier()
    
    qc.ccx(qr[7], qr[11], qr[12])
    qc.cx(qr[7],qr[11])
    qc.barrier()
    
    qc.append(mcx,[8,11,12,13])
    qc.ccx(qr[8], qr[11], qr[12])
    qc.cx(qr[8],qr[11])
    qc.barrier()
    
    qc.append(mcx,[9,11,12,13])
    qc.ccx(qr[9], qr[11], qr[12])
    qc.cx(qr[9],qr[11])
    qc.barrier()
    
    qc.x(qr[11])
    qc.x(qr[12])
    qc.barrier()
    
    #El qubit 14 indicará cuando se haya encontrado un estado deseado
    qc.append(mcx,[11,12,13,14])
    
    return qc
```

```python
def oracle(qc,qr):
    
    qc=oracle_part(qc,qr)
    qc1=qc.inverse()
    qc.z(14)
    qc.compose(qc1,inplace=True)
    
    return qc
```

```python
qr=QuantumRegister(15)
qc=QuantumCircuit(qr)
oracle(qc,qr).draw()
```

```python
def grover_diffusion(qc,qr):
    mcz = ZGate().control(4)
    
    qc.barrier()
    qc.h(range(5))
    qc.x(range(5))
    qc.append(mcz,list(range(0,5)))
    qc.x(range(5))
    qc.h(range(5))
    qc.barrier()
    return qc
```

```python
def grover(qc,qr):
    return grover_diffusion(oracle(qc,qr),qr)
```

```python
qr=QuantumRegister(15)
qc=QuantumCircuit(qr)

qra=QuantumRegister(15)
cra=ClassicalRegister(5)
qa=QuantumCircuit(qra,cra)

qa.h(range(5))
qb=grover(qc,qr)
qa.compose(qb.compose(qb),inplace=True)

qa.measure(range(5),cra)

qa.draw()
```

```python
qasm_simulator = Aer.get_backend('qasm_simulator')

job = execute(qa, qasm_simulator, shots = 1000)
counts = job.result().get_counts(qa)
print(counts)
plot_histogram(counts)
```

<h2>Usando Qiskit:</h2>

```python
def qft_rotations(circuit, n):
    """Performs qft on the first n qubits in circuit (without swaps)"""
    if n == 0:
        return circuit
    n -= 1
    circuit.h(n)
    for qubit in range(n):
        circuit.cp(np.pi/2**(n-qubit), qubit, n)
    # At the end of our function, we call the same function again on
    # the next qubits (we reduced n by one earlier in the function)
    qft_rotations(circuit, n)

# Let's see how it looks:
qc = QuantumCircuit(3)
qft_rotations(qc,3)
qc.draw()
```

```python
def swap_registers(circuit, n):
    for qubit in range(n//2):
        circuit.swap(qubit, n-qubit-1)
    return circuit

def qft(circuit, n):
    """QFT on the first n qubits in circuit"""
    qft_rotations(circuit, n)
    swap_registers(circuit, n)
    return circuit

qc = QuantumCircuit(3)
qc.x(2)
qc.barrier()
qft(qc,3)
qc.draw(output='mpl')

```

```python
ket_system = Statevector(qc)
ket_system.draw('Latex')
```
